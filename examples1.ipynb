```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dominance = pd.read_csv("Gachomba_et_al_2022/dominance_BLto5.csv", sep=';')
dominance
```

```python
dominance['baselines'].str.split(',', expand=True)
```

```python
# split by condition (in this case, dominant (C1) VS submissive (C2))
C1_dict = {}
C2_dict = {}
cols = ['baselines', 'd1', 'd2', 'd3', 'd4', 'd5']#, 'avg5']
for column in dominance.columns:
    if column in cols:
        splits = dominance[column].str.split(',', expand=True)
        C1_dict.update({column: splits[0].astype(float)}) 
        C2_dict.update({column: splits[1].astype(float)})

C1 = pd.DataFrame(C1_dict)
C2 = pd.DataFrame(C2_dict)
C1
```

```python
C1.mean()
```

```python
# using numpy / matplotlib
C1.to_numpy().mean(0)
```

```python
C1_np = C1.to_numpy()
C2_np = C2.to_numpy()
```

```python
x = [0, 1, 2, 3, 4, 5]
x_labels = ['baselines', 'd1', 'd2', 'd3', 'd4', 'd5']
fig, ax = plt.subplots()
ax.errorbar(x, C1_np.mean(0), yerr=np.std(C1_np, ddof=1) / np.sqrt(np.size(C1_np)), marker='o')
ax.errorbar(x, np.nanmean(C2_np, 0), yerr=np.nanstd(C2_np, 0, ddof=1) / np.sqrt(np.size(C2_np)), marker='o')
ax.set_xticks(x)
ax.set_xticklabels(x_labels)
ax.set_xlabel("sessions")
ax.set_ylabel("Prosocial choice (%)")
ax.set_ylim([30, 90])
```

```python
# using pandas / seaborn / scipy
from scipy.stats import sem
ax = C1.plot(yerr=sem(C1_np, 0))
```

```python
print("(different mean) t-statistic = %6.4f / p-value = %6.4f" % st.ttest_ind(C1['d1'], C2['d1'], nan_policy='omit'))
```

```python
print("(different) D = %6.4f / p-value = %6.4f" % st.ks_2samp(C1['d1'], C1['d2']))
```
